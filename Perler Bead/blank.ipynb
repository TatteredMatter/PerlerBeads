```python
location_of_different_colors = [[0,0]]
list_of_different_colors = im_array[[0], [0]]
counter = 1

#loop through each row
for row in range(len(im_array)):
    
    #loop through item in a row
    for item_array in range(len(im_array[row])):

        #save the array item to compare
        a = im_array[[row], [item_array]]
        #assume this item will be a new color
        new_color = True

        #loop through all already found colors
        for possible_colors in location_of_different_colors:
            # save the currect color from the list to compare
            b = im_array[[possible_colors[0]], [possible_colors[1]]]

            #compare the current array item to the current list item
            if (a == b).all():
                #if they're the same it isn't a new color
                new_color = False
        
        #if you've gotten through the list and the array item hasn't matched any previous items
        #then it is a new color
        if new_color == True:
            #increase the total count of colors
            counter += 1
            #add the color location to the location array
            location_of_different_colors.append([row,item_array])
            #add the color type to the array of all distinct colors
            list_of_different_colors = np.append(list_of_different_colors, im_array[[row], [item_array]], axis=0)

print(str(counter) + " different colors.")
list_of_different_colors
```

```python
df_all_pixels_by_row = df_all_pixels.head(64)
df2 = df_all_pixels.iloc[64:].reset_index(drop=True)
count = 0

while count < 63:
    count += 1
    df_all_pixels_by_row = df_all_pixels_by_row.join(df2, lsuffix='', rsuffix='_' + str(count))
    df2 = df2.iloc[64:].reset_index(drop=True)

df_all_pixels_by_row
```

```python
# Filter out the rows where all the values are the same
df_all_pixels_by_row_unique = df_all_pixels_by_row[df_all_pixels_by_row.nunique(axis=1).ne(1)]

df_all_pixels_by_row_unique
```

```python
# Filter out the columns where all the values are the same
df_all_pixels_by_column_unique = df_all_pixels_by_row_unique

if len(df_all_pixels_by_row_unique.groupby(['R', 'G', 'B'])) == 1:
    print("Deleting First Column")
    df_all_pixels_by_column_unique = df_all_pixels_by_column_unique.drop(['R', 'G', 'B'], axis=1)

count = 0

while count < 63:
    count += 1
    
    distinct_colors = len(df_all_pixels_by_column_unique.groupby(['R_' + str(count), 'G_' + str(count), 'B_' + str(count)]))
    if distinct_colors == 1:
        print("deleting column " + str(count))
        df_all_pixels_by_column_unique = df_all_pixels_by_column_unique.drop(['R_' + str(count), 'G_' + str(count), 'B_' + str(count)], axis=1)

df_all_pixels_by_column_unique
```
